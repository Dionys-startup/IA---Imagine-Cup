<h3> Extração das informações e manipulação dos dados

```python
import pandas as pd
import numpy as np
```

```python
estabelecimentos = pd.read_excel('Lugares_atualizados.xlsx', sheet_name='Estabelecimentos', index_col=0)
estabelecimentos
```

```python
for preco in estabelecimentos['Preco']:
    if preco >= 0.0 and preco <= 40.0:
        estabelecimentos['Preco'] = estabelecimentos['Preco'].replace(preco, 1)
    elif preco > 40.0 and preco <= 90.0:
        estabelecimentos['Preco'] = estabelecimentos['Preco'].replace(preco, 2)
    elif preco > 90.0 and preco <= 140.0:
        estabelecimentos['Preco'] = estabelecimentos['Preco'].replace(preco, 3)
    elif preco > 140.0 and preco <= 200.0:
        estabelecimentos['Preco'] = estabelecimentos['Preco'].replace(preco, 4)
    else:
        estabelecimentos['Preco'] = estabelecimentos['Preco'].replace(preco, 5)
        
estabelecimentos
```

```python
clientes = pd.read_excel('Lugares_atualizados.xlsx', sheet_name='Clientes', index_col=0)
clientes.head()
```

```python
price = pd.read_excel('Lugares_atualizados.xlsx', sheet_name='Range_Preços')
price.head()
```

<h3> Vetorização dos estabelecimentos

```python
from sklearn.feature_extraction import DictVectorizer

dv_1 = DictVectorizer(sparse=False)
dv_estabelecimentos = dv_1.fit_transform(estabelecimentos.apply(dict, 1))

print(dv_estabelecimentos)
```

```python
df_estabelecimentos = pd.DataFrame(dv_estabelecimentos, columns=dv_1.get_feature_names_out())
df_estabelecimentos.head()
```

<h3> Vetorização dos clientes

```python
dv_2 = DictVectorizer(sparse=False)
dv_clientes = dv_2.fit_transform(clientes.apply(dict, 1))

print(dv_clientes)
```

```python
df_clientes = pd.DataFrame(dv_clientes, columns=dv_2.get_feature_names_out())
df_clientes.head()
```

<h3> Agrupamento de colunas da tabela de estabelecimentos

```python
from sklearn.decomposition import PCA

estilo_columns = [col for col in df_estabelecimentos.columns if 'Estilo Musical' in col]

df_estilo = df_estabelecimentos[estilo_columns]

pca = PCA(n_components=1)

pca.fit(df_estilo)

df_estilo_compact = pca.transform(df_estilo)

df_estabelecimentos['estilo_compact'] = df_estilo_compact

df_estabelecimentos = df_estabelecimentos.drop(estilo_columns, axis=1)

print(df_estabelecimentos)
```

```python
local_columns = [col for col in df_estabelecimentos.columns if 'Local' in col]

df_local = df_estabelecimentos[local_columns]

pca = PCA(n_components=1)

pca.fit(df_local)

df_local_compact = pca.transform(df_local)

df_estabelecimentos['local_compact'] = df_local_compact

df_estabelecimentos = df_estabelecimentos.drop(local_columns, axis=1)

print(df_estabelecimentos)
```

<h3> Agrupamento de colunas da tabela de clientes

```python
estilo_columns = [col for col in df_clientes.columns if 'Estilo Musical' in col]

df_estilo = df_clientes[estilo_columns]

pca = PCA(n_components=1)

pca.fit(df_estilo)

df_estilo_compact = pca.transform(df_estilo)

df_clientes['estilo_compact'] = df_estilo_compact

df_clientes = df_clientes.drop(estilo_columns, axis=1)

print(df_clientes)
```

```python
zona_columns = [col for col in df_clientes.columns if 'Zona' in col]

# Create a subset of the dataframe with only the estilo_columns
df_zona = df_clientes[zona_columns]

# Create the PCA model
pca = PCA(n_components=1)

# Fit the PCA model to the estilo_columns
pca.fit(df_zona)

# Transform the estilo_columns using the PCA model
df_zona_compact = pca.transform(df_zona)

# Add the compacted columns to the original dataframe
df_clientes['local_compact'] = df_zona_compact

# Drop the estilo_columns from the dataframe
df_clientes = df_clientes.drop(zona_columns, axis=1)

# Print the updated dataframe
print(df_clientes)
```

<h3> Cálculo das similaridades entre os cossenos

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities = []

for index, row in df_clientes.iterrows():
    row_similarities = []
    for _, estabelecimento_row in df_estabelecimentos.iterrows():
        cos_sim = cosine_similarity([row.values], [estabelecimento_row.values])
        row_similarities.append(cos_sim[0][0])
    similarities.append(row_similarities)

similarities
```

```python
i = 0
melhor_escolha = []

for vetor in similarities:
   min = np.argmin(vetor)
   melhor_escolha.append(min)
   i += 1

print(melhor_escolha)
```
